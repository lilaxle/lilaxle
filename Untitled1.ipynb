```python
# 1 Import
import pandas as pd

# 2 Load data
data = pd.read_csv("https://raw.githubusercontent.com/owid/co2-data/master/owid-co2-data.csv")

# 3 Explore
print("Shape:", data.shape)
print("Columns:", list(data.columns))
print("First rows:")
print(data.head())

# 4 Manipulate
data["co2_per_capita"] = data["co2"] / data["population"]

# 5 Mini challenge
# Highest in 2020
data_2020 = data[data['year'] == 2020]
highest = data_2020.nlargest(1, 'co2_per_capita')
print("Highest CO2 per capita 2020:", highest['country'].values[0])

# US change
ro = data[data['country'] == 'United States']
ro_2000 = ro[ro['year'] == 2000]['co2_per_capita'].values[0]
ro_2020 = ro[ro['year'] == 2020]['co2_per_capita'].values[0]
change = ((ro_2020 - ro_2000) / ro_2000) * 100
print(f"US change 2000-2020: {change:.1f}%")

# 6. Cleaning
print("Missing values:", data.isnull().sum().sum())
data = data.fillna(0)
data = data.drop_duplicates()
print("After cleaning - Missing:", data.isnull().sum().sum(), "Duplicates:", data.duplicated().sum())

# 7. Reflection
print("Many countries, many cells, big differences in CO2 per capita")
```
